##  5.3.2 the mean of a random variable

### 5.3.2a formula 
- random variables are used to model population
- $\mu$ is used for population mean.
$$E(X) = \mu = \mu _ X = \frac{x_1 + x_2 + ··· x_n}{n} = \frac{1}{n}\sum_{i=1}^{n}{n(\{X=x_i\})}= ···= \sum_{\textrm{all x}}{x·p(x)}$$
- the capital E stands for *expected value* and is a function 

## 5.3.2b rules
assume a and b are fixed numbers and X and Y are random variables
1. (linearity rule for means) if X is a random variable and a and b are fixed numbers then 
$$\mu_{a+bX} = a + b\mu_X$$

you can take the constant out
2. (addition rule for means)
$$\mu_{X \pm Y} = \mu_X + \mu_Y$$

rule 1 and 2 are often combined

3. (for functions of a random variable that are not linear)if X is a random variable and g is a function of X, then: 
$$E(g(X)) = \sum{g(x_i)p_i}$$

## 5.4 Binomial and Poisson distribution
### binomial distribution: mean and standard deviation
if X ~ B(n,P) then

$$ E(X) = \mu_X=np$$
$$ \sigma_X=\sqrt{np(1-p)}$$

### Poisson distribution
$$p(x) = P(X = x)  = \frac{e^{\lambda}\lambda^x}{x!}$$
for x = 0,1,2,... and lambda > 0

$$\mu_X=\sigma^2=\lambda$$
$$\sigma_X=\sqrt{\lambda}$$
